```python
import pandas as pd
```

```python
df = pd.read_csv("HR Data.csv")
df.head(5)
```

```python
print("Initial Data:")
print(df.head(5))
```

```python
print("DataFrame Info:")
print(df.info())
```

```python
print("Missing Values per Column:")
print(df.isnull().sum())
```

```python
columns_to_remove = ['EmployeeCount', 'Over18', 'StandardHours']
df.drop(columns=columns_to_remove, inplace=True, errors='ignore')
df.head()
```

```python
df.rename(columns={
    'EmployeeNumber':'Employee_ID',
    'Attrition': 'Employee_Attrition',
    'MonthlyIncome': 'Monthly_Income',
    },inplace=False)
df.head()
```

```python
df.drop_duplicates(subset=['Employee_ID'], inplace=True)
df.head()
```

```python
String_columns = df.select_dtypes(include=['object']).columns
df[String_columns]=df[String_columns].apply(lambda x: x.str.strip())
df.head()
```

```python
df.dropna(inplace=True)
df.head()
```

```python
print('Missing values per column after cleanup:')
print(df.isnull().sum())
```

```python
df['Employee_Attrition']=df['Employee_Attrition'].astype('category')
df.info()
```

```python
print("Summary Statistics of cleaning:")
print(df.describe(include='all'))
```
